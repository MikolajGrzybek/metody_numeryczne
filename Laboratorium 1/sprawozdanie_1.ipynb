***Zadanie 1.***  
Zaimplementuj funkcj $cylinder\_area$ tak by zwracaa pole powierzchni walca o promieniu podstawy $r$ i wysokoci $h$. Staa $\pi$ jest zdefiniowana np. w bibliotece *[math](https://docs.python.org/3/library/math.html#constants)*. Je偶eli nie da si policzy pola funkcja powinna zwraca warto *NaN*. Sprawd藕 dziaanie zaimplementowanej funkcji dla dowolnych wartoci w tym notatniku.

```python
def cylinder_area(r: float, h: float):
    if r > 0 and h > 0:
        bases_area = 2 * math.pi * math.pow(r, 2)
        wall_area = 2 * math.pi * r * h
    else:
        return np.nan
    return bases_area + wall_area
```

Do wykonania operacji odwracania macierzy nale偶y u偶y funkcji *[inv](https://docs.scipy.org/doc/numpy/reference/generated/numpy.linalg.inv.html)*, do obliczenia wyznacznika macierzy stosuje si funkcj *[det](https://docs.scipy.org/doc/numpy-1.15.0/reference/generated/numpy.linalg.det.html)*  z bliblioteki *[linalg](https://docs.scipy.org/doc/numpy/reference/routines.linalg.html)*. Natomiast transpozycj macierzy wykonujemy przez funkcj *[transpose](https://docs.scipy.org/doc/numpy-1.14.0/reference/generated/numpy.transpose.html)* kt贸rej skrucona wersja to $M.T$ z pakietu *numpy*.  
    
*Ciekawostka:* Python natywnie nie zawiera struktury danych typu tablica, na poziomie jzyka jest to rozwizane poprzez struktur listy list do kt贸rej element贸w odwouje si poprzez indeksy jak w C/C++.

```python
def  ○_○(a: float):
    M = np.array([[2*a, 1, -a],[0, 1, 1],[-a, a, 1]])

    M_det = np.linalg.det(M)
    
    if(M_det == 0):
        M_trnp = np.nan
    else:
        M_trnp = np.transpose(M)

    M_inv = np.linalg.inv(M)
    
    print("Macierz:\n {0}".format(M))
    print("Macierz odwr贸cona:\n {0}".format(M_inv))
    print("Macierz transponowana:\n {0}".format(M_trnp))
    
    return (M_inv, M_trnp, M_det)
    
(mi, mt, md) =  ○_○(10)
```

***Wskaz贸wka:*** Inicjalizacja pustej macierz wykonywana jest w pakiecie Numpy przy pomocy funkcji  *[zeros](https://docs.scipy.org/doc/numpy-1.15.0/reference/generated/numpy.zeros.html)*, za macierzy skadajcej si z jedynek *[ones](https://docs.scipy.org/doc/numpy/reference/generated/numpy.ones.html)*.

```python
zero_matrix = np.zeros((2, 2))
ones_matrix = np.ones((3,1))

print('zero_matrix: \n{}'.format(zero_matrix))
print('ones_matrix: \n{}'.format(ones_matrix))
```

```python
def custom_matrix(m, n):
    
    if n < 0 or m < 0:
        return None
    
    M = np.zeros((m,n))
    # m - wiersz
    # n - kolumna
                 
    for x in range(0, n):
        for i in range(0, m):
            if x > i:
                M[x, i] = x
            else:
                M[x, i] = i
                
    print("{0}".format(M))
    return



custom_matrix(4,4)
                
                
                
```
